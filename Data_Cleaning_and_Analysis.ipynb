- The purchase price column is derived from the purchase_price table, which provides product-wise actual and purchases prices. The combination of vendor and brand is unique in this table.
- The vendor_invoice table aggregates data from the purchase table, summarising quantity and dollar amounts, along with an additional column for freight. This table maintains uniqueness based on vendor and PO Number.

--------------------------------------------------------------------------------------------------------------------------------------------------

As the data we need to address the objectives of the project is distributed in different tables, we need to create a summary table containing:
- purchases transactions made by vendors
- sales transaction data
- freight costs for each vendor
- actual product prices from vendor

```python
vendor_invoice.columns
```

```python
freight_summary = vendor_invoice.groupby('VendorNumber')['Freight'].sum().reset_index()
freight_summary.rename(columns={'Freight': 'FreightCost'}, inplace=True)
```

```python
freight_summary
```

```python
purchase_prices.columns
```

```python
p = purchases
pp = purchase_prices

p_selected = p[['VendorNumber', 'VendorName', 'Brand', 'Description', 'PurchasePrice', 'Quantity', 'Dollars']]

pp_selected = pp[['Brand', 'Volume', 'Price']]

merged_df = pd.merge(p_selected, pp_selected, on='Brand')

purchase_summary = (merged_df.groupby(['VendorNumber', 'VendorName', 'Brand', 'Description'], as_index=False).agg(
        TotalPurchaseQuantity=('Quantity', 'sum'),
        TotalPurchaseDollars=('Dollars', 'sum'),
        PurchasePrice=('PurchasePrice', 'first'),
        Volume=('Volume', 'first'),
        ActualPrice=('Price', 'first')))
```

```python
purchase_summary
```

```python
sales.columns
```

```python
sales_summary = (sales.groupby(["VendorNo", "Brand"], as_index=False).agg(
        TotalSalesDollars=("SalesDollars", "sum"),
        TotalSalesPrice=("SalesPrice", "sum"),
        TotalSalesQuantity=("SalesQuantity", "sum"),
        TotalExciseTax=("ExciseTax", "sum")).sort_values("TotalSalesDollars"))
```

```python
sales_summary
```

```python
# Merge purchase_summary with sales_summary
merged_df = purchase_summary.merge(
    sales_summary,
    left_on=["VendorNumber", "Brand"],
    right_on=["VendorNo", "Brand"],
    how="left"
)

# Merge the result with freight_summary
vendor_sales_summary = merged_df.merge(
    freight_summary,
    on="VendorNumber",
    how="left"
)

# Select and order columns like in the SQL
vendor_sales_summary = vendor_sales_summary[
    [
        "VendorNumber",
        "VendorName",
        "Brand",
        "Description",
        "PurchasePrice",
        "ActualPrice",
        "Volume",
        "TotalPurchaseQuantity",
        "TotalPurchaseDollars",
        "TotalSalesQuantity",
        "TotalSalesDollars",
        "TotalSalesPrice",
        "TotalExciseTax",
        "FreightCost"
    ]
]

# Sort by TotalPurchaseDollars (descending)
vendor_sales_summary = vendor_sales_summary.sort_values(
    by="TotalPurchaseDollars", ascending=False
)

vendor_sales_summary.reset_index(drop=True, inplace=True)
```

```python
vendor_sales_summary['Volume'] = vendor_sales_summary['Volume'].astype('float64')
```

```python
vendor_sales_summary.fillna(0, inplace = True)
```

```python
vendor_sales_summary['VendorName'] = vendor_sales_summary['VendorName'].str.strip()
```

```python
vendor_sales_summary['GrossProfit'] = vendor_sales_summary['TotalSalesDollars'] - vendor_sales_summary['TotalPurchaseDollars']
```

```python
vendor_sales_summary['ProfitMargin'] = (vendor_sales_summary['GrossProfit'] / vendor_sales_summary['TotalSalesDollars']) * 100
```

```python
vendor_sales_summary['StockTurnover'] = vendor_sales_summary['TotalSalesQuantity'] / vendor_sales_summary['TotalPurchaseQuantity']
```

```python
vendor_sales_summary['SalesToPurchaseRatio'] = vendor_sales_summary['TotalSalesDollars'] / vendor_sales_summary['TotalPurchaseDollars']
```
